```python
class Account:
    def __init__(self, frist_name   =input('Enter Your Frist Name ='),
                       last_name    = input('Enter Your Last Name ='),
                       city         = input('Enter you city ='),
                       user_id      = input('Enter Your user id ='),
                       password     = input('Enter Your Password ='),
                       phone_number = int(input('Enter your Phone number ='))):
        
        self.frist_name = frist_name
        self.last_name = last_name
        self.city      = city
        self.user_id   = user_id
        self.password  = password
        self.phone_number = phone_number
    
    def login(self):
        while True:
            sign_in = input('Enter your email')
            password = input('Enter your password')
            if sign_in ==self.user_id and password ==self.password:
                print('Sign in Sussesfully')
                break
            else:
                print('username or password is incorrect')
                continue
            
```

```python
class User_Account(Account):
    def __init__(self,acc_num,balance,*args):
        super(User_Account,self).__init__(*args)
        self.acc_num = acc_num
        self.balance = balance
        
        
    def user_info(self):
        while True:
            account = int(input("Enter Your Account Number"))
            if account == self.acc_num:
                print("Your Name :",self.frist_name,self.last_name)
                print("Your Phone Number :",self.phone_number)
                print("Your account number :",self.acc_num)
                print("Your account balance :",self.balance)
                break
            else:
                print("Username or Password is invalid")
                
    def deposite(self):
        print("Your Account Balance is :",self.balance)
        add_cash=int(input("Enter the Amount"))
        self.balance = add_cash+ self.balance
        print("You bank has been credited INR ",add_cash,"to your account ",self.acc_num,
              "and your current balance is ",self.balance)
        
    def transfer(self):
        print("Your Account Balance is :",self.balance)
        withdrawl = int(input("Enter the Amount you want to withdraw"))
        if withdrawl <=self.balance:
            self.balance = self.balance - withdrawl
            print("Your account is debited with ",withdrawl, 'against account number',self.acc_num,
                  '.Your current balance is ',self.balance)
            
        else:
            print("Opps seems your account do not have sufficent balance")
            
```

```python
class Tyre:
    def __init__(self,size,belted_biaz,pressure):
        self.size = size
        self.belted_biaz= belted_biaz
        self.pressure = pressure
        
    def __str__(self):
        return("Tyre : \n \tsize:"+str(self.size) +
              "\n \tbelted-biaz:"+(strself.belted_biaz)+
               "\n \tpressure:"+str(self.pressure))
    
class Engine:
    def __init__(self,noise,engine_type,effiency):
        self.noise = noise
        self.engine_type = engine_type
        self.effiency = effiency
        
    def __str__(self):
        return("Engine : \n \tnoise:"+ str(self.noise)+
               "\n \tEngine-Type:"+(self.engine_type)+
               "\n \teffiency"+str(self.effiency) )
class Body:
    def __init__(self,size,weight):
        self.size = size
        self.weight = weight
        
    def __str__(self):
        return("Body : \n \tsize:"+str(self.size)+
               "\n \tweight:"+(str(self.weight))
    
class Car:
    def __init__(self,tyre,engine,body):
        self.tyre = tyre
        self.engine = engine
        self.body = body
        
    def __str__(self):
        return str(self.tyre)+ '\n' + str(self.engine)+ '\n' +str(self.body)
    
t = Tyre(30,'Yes',35)
e = Engine(20,'Petrol',23.5)
b = Body(44,290)

c = Car(t,e,b)
print(c)
```

# Problem 1.

There are 1000 students who's marks are captured based on their ability in the test. if the same test is given to a new student what is the probibility of getting 50 marks.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from numpy.random import randint as ri
from pandas.plotting import scatter_matrix
import warnings
warnings.filterwarnings('ignore')
```

```python
marks = ri(30,101,1000)
df = pd.DataFrame(marks)
df.head()
```

```python
import scipy.stats as st
df_mean = df.mean()
print('Mean', df_mean)


df_std = df.std()
print('Standard_Dev', df_std)
```

```python
x=50
```

```python
prob = st.norm.cdf(x,df_mean,df_std)
print(prob)
```

A car manufacture company claims mileage of all their cars is 25kmpl. The population standard deveation is known of for the past experience to be 1.5 kmpl. A car rental company wants to test the null hypothesis that the average mileage is atlest 25kmpl. Car rental representative drove 30 (sample) of the same company and recorded each car mileage. The average mileage of the sample turned out to be 24.2kmpl.

Null Hypothesis >= 25kmpl
alternative Hypothesis = < 25kmpl

```python
mean = 25
std = 1.5
sample_mean = 24.2
n= 30
```

```python
# for p value we have to find sample standard deviation formula for that is 

# standard deviation of population/squareroot(sample polulation) 1.5/np.sqrt(30)
```

```python
import scipy.stats as st
import numpy as np
p_value = st.norm.cdf(sample_mean,mean,scale=std/np.sqrt(30))
p_value
```

# Anova (Analysis of Variance)

if we have more than 2 samples to check then anova comes into picture.

```python
smster_mrks = pd.read_csv('https://raw.githubusercontent.com/training-ml/Files/main/Semester%20marks.csv')
smster_mrks.head()
```

```python
import statsmodels.formula.api as sm
```

```python
dm = sm.ols('Total_Marks ~ Semester', data=smster_mrks).fit()
dm
```

```python
import statsmodels.api as k

anov_mod = k.stats.anova_lm(dm)
anov_mod
```

```python
# the p value is 0.749102 which is grater then alpha value it means we fail to refect null hypothesis
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
admission_pre = pd.read_csv('https://raw.githubusercontent.com/training-ml/Files/main/Admission_Prediction.csv')
admission_pre.head()
```

```python
admission_pre.shape
```

```python
#in this we need to check the statistics of data and write our observation
admission_pre.describe()
```

Need to fill null Values in  TOEFL Score, University Rating, GRE Score

```python
admission_pre['University Rating'] = admission_pre['University Rating'].fillna(admission_pre['University Rating'].mode()[0])

admission_pre['GRE Score'] = admission_pre['GRE Score'].fillna(admission_pre['GRE Score'].mean())

admission_pre['TOEFL Score'] = admission_pre['TOEFL Score'].fillna(admission_pre['TOEFL Score'].mean())


```

```python
admission_pre.describe()
```

```python
df1 = admission_pre.drop(columns=['Serial No.'])
```

```python
df1.head()
```

```python
plt.figure(figsize=(15,10))

plotnumber =1

for columns in df1:
    if plotnumber<=8:
        ax = plt.subplot(3,3,plotnumber)
        sns.distplot(df1[columns])
        plt.xlabel(columns,fontsize =20)
    plotnumber+=1
    
plt.show()    
        
```

All continous data is normally distibuted so we can go ahead.

```python
x = df1.drop(columns=['Chance of Admit'])
y = df1['Chance of Admit']
```

Now we divided features and lable serpatley

```python
plt.figure(figsize =(15,10), facecolor = 'Yellow')
plotnumber=1

for column in x:
    if plotnumber<=8:
        ax = plt.subplot(3,3,plotnumber)
        plt.scatter(x[column],y)
        plt.xlabel(column)
        plt.ylabel('Chance of Admit')
    plotnumber+=1 
plt.show()    
```

As we can see postive relationship between feature and lable

```python
scaler = StandardScaler()

x_scaled = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.20,random_state = 340)
```

```python
reg_model = LinearRegression()

reg_model.fit(x_train,y_train)
```

```python
admission_pre.head(2)
```

```python
print('admission Prediction :' , reg_model.predict(scaler.transform([[337.0,118.0,4.0,4.5,4.5,9.65,1]])))
```

```python
reg_model.score(x_train,y_train)
```

```python
reg_model.score(x_test,y_test)
```

# Model Evaluation

```python
y_pred = reg_model.predict(x_test)
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel("Actual chance of admission")
plt.ylabel('Predicted chance of admission')
plt.title('Actual vs Prediction')
plt.show()
```

# Regularzation

```python
# frist we have to find learning rate(alpha)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge,Lasso,RidgeCV,LassoCV
```

```python
lassocv = LassoCV(alphas = None, max_iter=1000, normalize= True )
lassocv.fit(x_train,y_train)
```

```python
alpha = lassocv.alpha_
alpha
```

```python
lasso_reg = Lasso(alpha)
lasso_reg.fit(x_train,y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

```python
# its trun for RidgeCV to train the model
```

```python
ridgecv = RidgeCV(alphas = np.arange(0.001,0.01,0.001), normalize = True)
ridgecv.fit(x_train,y_train)
```

```python
alphas = ridgecv.alpha_
alphas
```

```python
ridgecv_reg = RidgeCV(alphas)
ridgecv_reg.fit(x_train,y_train)
```
